```python
#captures frames from video into images
import cv2
import numpy as np
 
# Opens the Video file
cap = cv2.VideoCapture("vid/vid3.mp4")
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == False:
         break
   
    cv2.imwrite('vid/img_vid3/img'+str(i)+'.jpg',frame)
    i+=1
 
cap.release()
cv2.destroyAllWindows()
```

```python
#deep learning test
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

### INPUT NODE
img_inputs = keras.Input(shape=(352, 640, 3))

### LAYERS
dense = layers.Dense(64, activation="relu")
x = dense(img_inputs)
x = layers.Dense(64, activation="relu")(x)
outputs = layers.Dense(10)(x)

### CREATE MODEL
model = keras.Model(inputs=img_inputs, outputs=outputs, name="mnist_model")

### MODEL SUMMARY
model.summary()

### PLOT GRAPH
keras.utils.plot_model(model, "my_first_model.png")

### PLOT GRAPH WITH INPUTS AND OUTPUTS
keras.utils.plot_model(model, "my_first_model_with_shape_info.png", show_shapes=True)
```

```python
#deep learning model
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

### INPUT NODE
inputs = keras.Input(shape=(352, 640, 3))

### DEFINE MODEL

## batch norminalisation

## convolution layer 1
x = tf.keras.layers.Conv2D(32, kernel_size=3, activation='relu')(inputs)

## convolution layer 2
x = tf.keras.layers.Conv2D(16, kernel_size=3, activation='relu')(x)

## flatten
x = tf.keras.layers.Flatten(data_format=None)(x)

## dense layer 1
x = tf.keras.layers.Dense(8, activation='relu')(x)

## dropout layer
x = tf.keras.layers.Dropout(.2)(x)

## dense layer 2
x = tf.keras.layers.Dense(4, activation='relu')(x)

### OUTPUT NODE
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

### CREATE MODEL
model = tf.keras.Model(inputs=inputs, outputs=outputs, name="klux_model")

### MODEL SUMMARY
model.summary()

### PLOT GRAPH WITH INPUTS AND OUTPUTS
tf.keras.utils.plot_model(model, "my_first_model_with_shape_info.png", show_shapes=True)
```
